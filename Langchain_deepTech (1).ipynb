```python
##What is langchain 
Langchain is a framework that help developers applications using large language models like GPT.
Instead of asking on question, Langchain allows you to connect multipke steps toghter like prompts, memory,tools,And API's create more and powerful systems



In simple terms, LangChain acts like a bridge between raw AI models and real-world applications.
It helps you organize how inputs are processed, how responses are generated, and how external data or tools are used.
```

```python
##Why is LangChain Important?

In real-world applications, we don’t just need one response from an AI — we need a flow of actions.
For example, a chatbot may need to remember previous conversations, search for information, and then generate a reply.

LangChain makes this possible by providing a structured way to manage all these steps.
It saves time, reduces complexity, and allows developers to focus more on building useful applications instead of handling everything manually.
```

```python
##Problems LangChain Solves
1. Managing Complex Workflows
Without LangChain, developers have to manually handle multiple steps like input processing, API calls, and output formatting. 
LangChain simplifies this by organizing everything into chains.

2. Connecting Tools with AI
LLMs alone cannot perform tasks like calculations or accessing live data.
LangChain allows integration with tools like calculators, search engines, and databases.

3. Adding Memory to AI
Normally, AI forgets previous interactions. 
LangChain introduces memory, so applications can remember past conversations and provide more relevant responses.

4. Handling Prompts Efficiently
Instead of writing prompts again and again, LangChain provides templates to reuse and manage prompts easily.
```

```python
##What is LLM?

A Large Language Model (LLM) is an advanced type of artificial intelligence that is trained on huge amounts of text data.
It can understand human language and generate meaningful responses based on the input it receives.

LLMs are capable of performing tasks like answering questions, summarizing content, translating languages, and even writing code. 
They work by predicting the next word in a sentence, which allows them to generate natural and coherent text.

In simple terms, an LLM acts like a smart assistant that can understand what we say and respond in a human-like way.
```

```python
##Why are LLMs Important?

Large Language Models are important because they act as the core intelligence behind modern AI applications. 
They make it possible for machines to understand and respond to human language in a meaningful way.

LLMs are widely used in real-world applications such as chatbots, virtual assistants, content generation, and coding support tools.
They help automate tasks, improve productivity, and enhance user experience.

Without LLMs, most of today’s Generative AI systems would not be possible, as they provide the foundation for building intelligent and interactive applications.
```

```python
##What is a Chat Model?

A chat model is a type of Large Language Model that is specifically designed for conversations.
Unlike basic LLMs that take plain text input, chat models work with structured messages such as user input and system instructions.

This structure helps the model understand the context of a conversation and generate more relevant and natural responses.
Chat models are commonly used in applications like chatbots, virtual assistants, and customer support systems.

In simple terms, a chat model is an AI that can talk with users in a more interactive and conversational way, making the experience feel more human-like.
```

```python
##What is Memory?


Memory in LangChain allows the model to remember previous interactions. 
It helps maintain context in a conversation, so the AI can give more relevant and meaningful responses.

In simple terms, memory allows the AI to “remember” what was said earlier.
```

```python
##Why Memory is Important?

Without memory, the model treats every input as a new request.
Memory helps in building conversational AI systems like chatbots that can understand context and continue conversations naturally.
```

```python
import os
from langchain_groq import ChatGroq
from langchain_community.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

os.environ["GROQ_API_KEY"] = "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"

llm = ChatGroq(model="llama-3.1-8b-instant")

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

print(conversation.predict(input="Hi, my name is Shannu"))
print(conversation.predict(input="What is my name?"))
```

```python
import os
from langchain_groq import ChatGroq
# New import path for agents
from langchain.agents import initialize_agent, Tool 
from langchain_core.tools import Tool

# 1. API Key (Use your fresh key)
os.environ["GROQ_API_KEY"] = "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" 

# 2. Model (Using the active Llama 3.1 model)
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)

# 3. Define the tool
def calculator_tool(input_text):
    # Note: eval() can be risky, but fine for a local learning script!
    return str(eval(input_text))

tools = [
    Tool(
        name="Calculator",
        func=calculator_tool,
        description="Useful for solving math problems. Input should be a math expression like 25*4"
    )
]

# 4. Initialize Agent (The "ReAct" logic)
agent_executor = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True,
    allow_dangerous_tools=True # Required in newer versions for tools like eval
)

# 5. Run using .invoke()
response = agent_executor.invoke({"input": "What is 25 * 4?"})
print(response["output"])
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# Load document
loader = TextLoader("E:\china_beijing.docx")
documents = loader.load()

# Create embeddings
embeddings = HuggingFaceEmbeddings()

# Store in vector DB
db = FAISS.from_documents(documents, embeddings)

# Query
query = "What is AI?"
docs = db.similarity_search(query)

print(docs[0].page_content)
```

```python
Advantages of LangChain

LangChain provides a modular and flexible framework for building applications using large language models.
It allows developers to combine different components like prompts, chains, memory, and tools in a structured way.

One of the key advantages is rapid development.
Developers can quickly prototype and build AI applications without handling low-level details.

LangChain also supports integration with various APIs, databases, and tools, making it highly adaptable for real-world use cases.

Another benefit is reusability.
Components like prompt templates and chains can be reused across different applications, reducing development effort.


    
Limitations of LangChain

Despite its advantages, LangChain has some limitations. One major issue is latency, as multiple components in a chain can slow down the response time.

Debugging can also be challenging because the workflow involves multiple steps, making it harder to identify where errors occur.

Cost is another concern when using paid APIs, especially for large-scale applications.

Additionally, LangChain may not be necessary for simple tasks, where directly using an LLM would be more efficient.

When Not to Use LangChain

LangChain should not be used for very simple applications that require only a single LLM call, as it may add unnecessary complexity.

It is also not ideal when low latency is critical, or when system resources are limited.

Simple Summary

LangChain is a powerful framework for building complex AI applications, but it comes with trade-offs such as increased latency, cost, and debugging complexity.
```
